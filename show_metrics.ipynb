```python
import numpy as np
from numpy import load, save
import matplotlib.pyplot as plt
```

### EffitientNet_b1 results

```python
train_loss = load('./effNet_result/train_loss.npy')
test_loss = load('./effNet_result/val_loss.npy')

plt.plot(train_loss, 'b', label='Train Loss')
plt.plot(test_loss, 'r', label='Valid Loss')

plt.xlabel('Epochs')
plt.ylabel('Loss')

plt.legend()
plt.title('Train and Valid Loss Over Epochs')
plt.show()
```

```python
train_iou = load('./effNet_result/train_iou.npy')
test_iou = load('./effNet_result/val_iou.npy')

plt.plot(train_iou, 'b', label='Train Iou')
plt.plot(test_iou, 'r', label='Test Iou')

plt.xlabel('Epochs')
plt.ylabel('Mean IoU')

plt.legend()
plt.title('Train and Validation MeanIoU Over Epochs')
plt.show()

print('Overfitt!')
```
